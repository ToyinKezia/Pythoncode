```python
import pandas as pd
import numpy as np
```

```python
maindf = pd.read_csv("C:\\Users\\Admin\\Anothermain.csv")
```

```python
#calculating ZhaoLiu Trade Openness and Export-Import
maindf['ImpExp'] = maindf['Import_Value'] + maindf['Export_Value']
maindfExp = maindf['Export_Value'].sum()
maindfImp = maindf['Import_Value'].sum()
maindf['NewExportOpen'] = (43* maindf['Export_Value']**2) / (maindf['AGRGDP'] *  maindf['Export_Value'].sum())
maindf['NewImportOpen'] = (43*maindf['Import_Value']**2) / (maindf['AGRGDP'] * maindf['Import_Value'].sum())
```

```python
maindf['NewAGTradeExpOpenPart'] = ((maindf['Export_Value'] / (maindf['Import_Value'] + maindf['Export_Value'])) * maindf['NewExportOpen'])
maindf['NewAGTradeImpOpenPart'] = ((maindf['Import_Value'] / (maindf['Import_Value'] + maindf['Export_Value'])) * maindf['NewImportOpen'])
maindf['NewAGTradeOpen'] = maindf['NewAGTradeExpOpenPart'] + maindf['NewAGTradeImpOpenPart']
```

```python
maindf['OldExpImpRatio'] = (maindf['Export_Value'] / maindf['Import_Value'])*100
maindf['NewExpImpRatio'] = (maindf['NewExportOpen']/ maindf['NewImportOpen'])*100
maindf
```

```python
maindf = maindf.drop(columns = ['OldExpImpRatioSurplus_Deficit', 'NewExpImpRatioSurplus_Deficit', 'NewAG_Infra', 'TO_Infra', 'NewAG_Infra', 'NewTrade_GenInfra', 'OldTrade_GenInfra'] )
```

```python
def add_high_low_column(maindf, column_name):
    new_column_name = column_name + "Surplus_Deficit"
    maindf[new_column_name] = maindf[column_name].apply(lambda x: 'Deficit' if x < 95 else("Balance" if x >= 100 <=110 else "Surplus"))
    return maindf
maindf = add_high_low_column(maindf, 'OldExpImpRatio')
maindf = add_high_low_column(maindf, 'NewExpImpRatio')
maindf
```

```python
maindf['GenInfra_OldTrade'] = maindf['OldAGTradeOpen'] * maindf['GeneraInfra']
maindf['GenInfra_NewTrade'] = maindf['NewAGTradeOpen'] * maindf['GeneraInfra']
maindf['Log_OldTrade'] = maindf['OldAGTradeOpen'] * maindf['Logistics']
maindf['Log_NewTrade'] = maindf['NewAGTradeOpen'] * maindf['Logistics']
```
