# Calculating Traditional Trade Openness

```python
maindf['ImpExp'] = maindf['Import_Value'] + maindf['Export_Value']
maindf
```

```python
maindf['OldAGTradeOpen'] = maindf['ImpExp'] / maindf['AGRGDP']
```

```python
maindf['OldAGImportOpen'] = maindf['Import_Value'] / maindf['AGRGDP']
```

```python
maindf['OldAGExportOpen'] = maindf['Export_Value'] / maindf['AGRGDP']
```

```python
#OldExportRatio is the degree of dependence on export trade
#OldImportRatio is the degree of dependence on import trade
#OldTradeOpen is the degree of dependence on Agricultural Foreign trade, ratio of import-export volume and GDP
```

# Now Calculating New Trade Openness

```python
#summation of the ImportExport Column
maindfImpExp = maindf['ImpExp'].sum()
maindfGroup= maindf.groupby(['Country', 'Year'])[['Import_Value', 'Export_Value']].sum()
```

```python
#calculating Squalli Trade Openness 
maindf['SqualliAGTradeOpen'] = (44* (maindf['ImpExp']**2)) / (maindf ['AGRGDP'] * maindf['ImpExp'].sum())
#maindf['SqualliTradeGroup'] = (44* (maindf['ImpExp']**2)) / (maindf ['AGRGDP'] * maindfGroup)
```

```python
#calculating ZhaoLiu Trade Openness and Export-Import
maindfExp = maindf['Export_Value'].sum()
maindfImp = maindf['Import_Value'].sum()
maindf['NewExportOpen'] = (44* (maindf['Export_Value']**2)) / (maindf['AGRGDP'] *  maindf['Export_Value'].sum())
maindf['NewImportOpen'] = (44*(maindf['Import_Value']**2)) / (maindf['AGRGDP'] * maindf['Import_Value'].sum())
maindf['NewAGTradeOpen'] = ((maindf['Export_Value'] / maindf['ImpExp']) * maindf['NewExportOpen']) + ((maindf['NewImportOpen'] / maindf['ImpExp']) * maindf['NewImportOpen'])
```

```python
maindf['OldExpImpRatio'] = (maindf['Export_Value'] / maindf['Import_Value'])*100
maindf['NewExpImpRatio'] = (maindf['NewExportOpen']/ maindf['NewImportOpen'])*100
maindf
```

```python
maindf['TO_Infra'] = maindf['TradeOpen'] * maindf['Infrastructure']
```

```python
Agrictrade = maindf[['Country', 'Year', 'TradeOpen', 'OldAGImportOpen', 'OldAGExportOpen', 'NewExportOpen','NewImportOpen', 'NewAGTradeOpen', 'SqualliAGTradeOpen',
'OldExpImpRatio', 'NewExpImpRatio', 'OldAGTradeOpen']]
Agrictrade
```

```python
def add_high_low_column(maindf, column_name):
    new_column_name = column_name + "Surplus_Deficit"
    maindf[new_column_name] = maindf[column_name].apply(lambda x: "Surplus" if x > 100 else "Deficit")
    return maindf

maindf = add_high_low_column(maindf, 'OldExpImpRatio')
maindf = add_high_low_column(maindf, 'NewExpImpRatio')
```

```python
maindf.rename(columns={'OldExpImpRatioSurplus_Deficit':'Status'}, inplace = True)
maindf.rename(columns={'NewExpImpRatioSurplus_Deficit':'NewStatus'}, inplace =True)
```
